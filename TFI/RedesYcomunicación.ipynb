```python
import matplotlib.pyplot as plt
import networkx as nx

# 1. Crear un grafo dirigido (DiGraph) - ¡Esencial para que sea direccionado!
G = nx.DiGraph()

# 2. Definir los 8 nodos (del 0 al 7) y sus nombres en un diccionario
nombres_nodes = {
    0: "Nodo 1",
    1: "Nodo 2",
    2: "Nodo 3",
    3: "Nodo 4",
    4: "Nodo 5",
    5: "Nodo 6",
    6: "Nodo 7",
    7: "Nodo 8",
}

# Añadir los nodos al grafo
G.add_nodes_from(nombres_nodes.keys())
# Asignar el atributo de nombre a cada nodo
nx.set_node_attributes(G, nombres_nodes, "nombres")

# 3. Crear las conexiones direccionadas (Cada nodo habla con los 2 siguientes)
num_nodos = len(nombres_nodes)
for i in range(num_nodos):
    vecino1 = (i + 1) % num_nodos
    vecino2 = (i + 2) % num_nodos

    G.add_edge(i, vecino1)
    G.add_edge(i, vecino2)

# 4. Dibujar el grafo visualizando las etiquetas personalizadas
# Usamos circular_layout para que la red de 8 nodos se aprecie de forma limpia y simétrica
pos = nx.circular_layout(G)
etiquetas = nx.get_node_attributes(G, "nombres")

plt.figure(figsize=(7, 7))
nx.draw(
    G,
    pos,
    labels=etiquetas,
    with_labels=True,
    node_color="lightblue",
    node_size=2000,
    font_size=10,
    arrowsize=15,  # Tamaño de las flechas de dirección
    connectionstyle="arc3,rad=0.1",  # Curva las líneas para que las flechas de ida y vuelta no se tapen
)

plt.title("Red Direccionada de 8 Nodos (Cada uno con 2 salidas)")
plt.show()
```
